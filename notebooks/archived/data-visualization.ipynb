__Question 1: What were the most popular names in 2013, and how did their popularity change over the past 10 years?__

To get started with this question, we need to do some subsetting, which you are very familiar with by now.  Let's look specifically at males for this first question.  First and foremost, however, we need to figure out the most popular male baby names in 2013.  A few `pandas`/Python methods that you've learned in previous assignments can get this done.  

Notice what we are doing here - you can think of the line of code as a _chain_ of methods in which we are manipulating the `df` data frame in turn.  

* First, we subset the data frame for only those male records in 2013; 
* Then, we sort the data frame in descending order by count; 
* Then, we slice the data frame to get back the top 15 rows; 
* Finally, we ask `pandas` to generate an array of unique names from our subsetted and sorted data frame.  

```python
top_names = df[(df.year == 2013) & (df.sex == 'M')].sort('n', ascending = False)[0:15].name.unique()

top_names
```

`pandas` returns a NumPy array of the top 15 baby names in 2013 for boys.  We can then pass this array to the `.isin()` method to get back entries for all of those names since 2000, and calculate their frequency per 1000 records in the dataset.  

```python
sub1 = df[(df.name.isin(top_names)) & (df.year > 1999) & (df.sex == 'M')]

sub1['per1000'] = sub1.prop * 1000

sub1.head()
```

We are just about ready to visualize the data now.  There are multiple ways these data could be visualized; in this instance, we'll use a __heatmap__, which we discussed in class.  A heatmap is a grid of cells in which the shading of each cell is proportional to its value.  Generally, darker cells represent a greater value.  When applied to temporal data, it can be an effective way to show the variation of values for multiple data series over time.  

Heatmaps in `seaborn` take a wide-format data frame with the y-values in the index, the x-values as the columns, and the data values in the cells.  We will use the `.pivot()` method to reshape our data and produce this type of data frame, then pass the dataframe to the `heatmap()` function.  

```python
wide1 = sub1.pivot(index = 'name', columns = 'year', values = 'per1000')

sns.heatmap(wide1)
```

The plot looks nice by default; we can see some trends such as the ascension of Liam and Aiden and the relative descent of Michael and Jacob (although both of those names are still in the top 15, of course).  However, the chart could use some cleaning up.  It would be nice if it were larger; in addition, the x-axis tick values are bunched together, making them difficult to read.  

`seaborn` plots have many plot customization options built-in; you'll learn how to use a few later in the assignment.  `seaborn` plots, however, are also `matplotlib` objects, which is the core plotting library in Python.  In turn, you can use the wealth of functions available in `matplotlib` to modify your `seaborn` plots.  You'll learn a few of those methods in this assignment.  

Note the code below and what we are doing.  We'll import the `pyplot` module from `matplotlib` in the standard way as `plt`.  `pyplot` gives us access to many different plot customization functions.  We can set the figure size before calling the plotting function, then rotate the x-tick labels, remove the axis labels, and add a title to our chart.  Also, notice the arguments passed to `sns.heatmap()`.  The `annot` parameter allows us to annotate the heatmap with data values, and the `cmap` parameter allows us to adjust the colors.  It accepts all ColorBrewer palettes as well as the built-in `matplotlib` palettes.  

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10, 7))

sns.heatmap(wide1, annot = True, cmap = 'Greens')

plt.xticks(rotation = 45)
plt.ylabel("")
plt.xlabel("")
plt.title("Most popular baby names - Male (rate per 1000)")
```

__Question 2: How does the release of Disney movies influence the popularity of baby names?__

Baby names can sometimes be responsive to trends in popular culture.  For example, "Daenerys" showed up in the dataset for the first time in 2012, and 67 baby girls were named Daenerys in 2013!.  In this exercise, we'll examine how the release of Disney Princess movies relates to baby names.  As our data only go up to 2013, we can't measure the impact of "Elsa" - but we can take a look at some other examples.  

Let's examine trends in female baby names since 1980 for four Disney Princess names: Jasmine, Ariel, Belle, and Tiana.  

```python
princesses = ['Jasmine', 'Ariel', 'Belle', 'Tiana']

sub2 = df[(df.name.isin(princesses)) & (df.sex == 'F') & (df.year > 1979)]

sub2['per1000'] = sub2.prop * 1000

sub2.head()
```

Clearly, Jasmine was a popular name in the early 1980s prior to the release of Aladdin.  Tiana, Ariel, and Belle, however, were not as popular.  So how did their popularity shift over time?  

We'll make a line chart using the built-in `.plot()` data frame method in `pandas`.  Unless you tell the method otherwise, `pandas` will put the index on the x-axis, and plot lines with the columns.  As such, we'll need to reshape our long-form data frame to wide again, placing `year` in the index.  We'll then plot to take a look at the data.  

```python
wide2 = sub2.pivot(index = 'year', columns = 'name', values = 'per1000')

wide2.plot(figsize = (10, 7))
```

We can start to get a sense here of some "spikes" in the data - for example, a clear spike in babies named Ariel is evident after 1989, which is when _The Little Mermaid_ was released.  Additionally, Belle has not been a popular name over the past 30 years, independent of the release of _Beauty and the Beast_.  

However - how can we express this on the chart in clearer terms?  One way to accomplish this is through __annotation__, which refers to the placement of text on the plot to highlight particular data points.  Before doing this, let's figure out approximately what the values are for each princess name when its movie was released: 

```python
disney_yrs = {'Ariel': 1989, 
             'Belle': 1991, 
             'Jasmine': 1992, 
             'Tiana': 2009}

for princess, year in disney_yrs.items(): 
    print(princess + ': ' + str(wide2[princess].ix[year]))
```

I accomplished this with a little new Python code.  I've mentioned before the _dictionary_, a type of Python object enclosed in curly braces (`{}`) that can hold key-value pairs.  The _key_ comes before the colon, the _value_ comes after the colon, and each element of the dictionary is separated by a comma.  

In this case, our dictionary holds the name of the Disney princess, and the year that the corresponding film was released.  Dictionaries can be iterated through with `for` and the `.items()` method; in this example, `princess` represents the key in the loop, and `year` represents the value.  Using the `.ix[]` data frame method in the loop allows us to access the data value for each princess for its corresponding year, which we'll use to annotate the plot.  

Spend some time reading through the `plt.annotate()` code below so that you can understand it.  We're using a number of parameters here: 

* The annotation text is the first argument.  Python will interpret the string '\n' as a line break, which allows us to put the text on multiple lines.  
* The `xy` parameter refers to the data coordinates of the point we want to annotate, given that we've specified this with the `'data'` argument supplied to the `xycoords` parameter.  We'll use the year of the film release for the X value, and the data values we obtained above (approximately) for the Y value.  
* In this case, however, we don't want to put the text right on top of the lines themselves; as such, we can specify an offset, and connect our text with the data point with an arrow.  We use the `xytext` and `textcoords` parameters to do this; have a look at the plot and see where this puts the text.  The argument supplied to `arrowprops`, which are in the form of a `dict`, govern the appearance of the arrow.  

Annotation often takes iteration and patience to get it right.  Try changing some of the arguments in the `plt.annotate()` calls below and see how the text and arrows move around!

```python
wide2.plot(figsize = (10, 7))

plt.annotate('The Little \nMermaid', xy = (1989, 0.64), xycoords = 'data', 
            xytext = (1985, 0.64), textcoords = 'data', arrowprops = dict(arrowstyle = '->'), 
            fontsize = 10)

plt.annotate('Aladdin', xy = (1992, 5.23), xycoords = 'data', 
            xytext = (1990, 6.1), textcoords = 'data', arrowprops = dict(arrowstyle = '->'), 
            fontsize = 10)

plt.annotate('Beauty and \nthe Beast', xy = (1991, 0.01), xycoords = 'data', 
            xytext = (1990, 0.95), textcoords = 'data', arrowprops = dict(arrowstyle = '->'), 
            fontsize = 10)

plt.annotate('The Princess \nand the Frog', xy = (2009, 0.247), xycoords = 'data', 
            xytext = (2007, 1.6), textcoords = 'data', arrowprops = dict(arrowstyle = '->'), 
            fontsize = 10)

plt.ylabel("Female baby names per 1000", fontsize = 12)
plt.xlabel("")
plt.title("Influence of Disney movies on baby names", fontsize = 15)
plt.legend(title = "", fontsize = 12)
```

__Question 3: How have gender-neutral names shifted in popularity between male and female over time?__

For the third and final question, we'll be looking at how four gender-neutral names have shifted between male and female over time.  Let's produce a new data frame from our original data frame that subsets for four popular gender-neutral names: Jordan, Riley, Peyton, and Taylor.  We'll take rows for years 1960 and later, and fill NaN values with 0.  

```python
neutral_names = ['Jordan', 'Peyton', 'Taylor', 'Riley']
```

```python
n1 = df[(df.name.isin(neutral_names)) & (df.year > 1959)].fillna(0)

n1.head()
```

Currently, our dataset is in the original _long form_, which tends to work well with `seaborn`.  In class, we discussed how to make small multiples plots using `pandas`; we'll reproduce for `seaborn` here.  

Faceted plots are available by mapping `seaborn` plotting functions onto a grid with the `FacetGrid` function, or by using the `factorplot` function, which plots onto a FacetGrid by default.  In turn, `factorplot` is a nice way to make small multiples plots using familiar syntax.  Let's try plotting faceted line charts that show how counts for these names vary by gender over time: 

```python
sns.factorplot(data = n1, x = 'year', y = 'n', hue = 'sex', col = 'name')
```

We can start to get a sense of some of the variations here; Taylor is more popular among girls than boys, whereas the opposite is true for Jordan.  Given that all of the plots share the same y-axis, however, it is tough to tell for Riley and Peyton.  Let's make a few modifications to the plot to improve its clarity.  We will add a `col_wrap` argument to specify how many columns to create in our plot grid, change the markers to a smaller size, increase the size of the charts, and set different y axis scales for each plot.  Be cautious when using inconsistent axes across the charts, though - it can make comparisons across plots confusing, as they are scaled differently!  We can also change the colors with the argument supplied to `palette`; we'll use common colors to represent female and male babies here.  

Additionally, notice how we are assigning the plot to a variable called `chart`.  `seaborn` factor plots themselves have methods that you can use to modify the chart appearance; we'll use a couple below to improve the look of our axes.  

```python
sns.set_style('whitegrid')

chart = sns.factorplot(data = n1, x = 'year', y = 'n', hue = 'sex', col = 'name', palette = ['pink', 'skyblue'], 
                       col_wrap = 2, markers = '.', join = True, sharey = False, size = 5)

chart.set_xticklabels(step = 10)
chart.set_axis_labels('Year', 'Number of names')
```
