````python
%matplotlib inline
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1983)

sns.distplot(np.random.randn(1000), color = 'purple')
````

Visualizations can be saved by right-clicking in your web browser and choosing "Save image as"; however, you'll get better quality by exporting your images from Python using the `savefig` function in `matplotlib`.  For example: 

````python
np.random.seed(1983)

sns.distplot(np.random.randn(1000), color = 'purple')
plt.savefig('myplot.png')
````

This saves your figure as a PNG image called `myplot.png` in your Notebook's directory with a resolution of 300 DPI.  I can now embed my image in a Markdown cell using the same Markdown syntax that is used to link to websites, so long as it is preceded with an exclamation point. ```markdown
